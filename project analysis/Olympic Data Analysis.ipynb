```python
# Convert 'start_date' and 'end_date' to datetime format
games['start_date'] = pd.to_datetime(games['start_date'], errors='coerce')
games['end_date'] = pd.to_datetime(games['end_date'], errors='coerce')
```

```python
# Calculate the most frequent date for 'start_date' and 'end_date'
most_freq_start_date = games['start_date'].mode()[0]
most_freq_end_date = games['end_date'].mode()[0]

# Impute missing values for 'start_date' and 'end_date' with the most frequent date
games['start_date'].fillna(most_freq_start_date, inplace=True)
games['end_date'].fillna(most_freq_end_date, inplace=True)
```

```python
# Check if missing values are handled
games.isnull().sum()
```

```python
# Impute missing values for 'isHeld' with the most frequent value
most_freq_isHeld = games['isHeld'].mode()[0]
games['isHeld'].fillna(most_freq_isHeld, inplace=True)
```

```python
# Check if missing values are handled
games.isnull().sum()
```

Now, all missing values are handeled

### 1. What are the total number of athletes in the dataset?

### 3. Which sport has the highest number of events?

### 4. What is the total number of medals won by each country?

```python
total_medals_per_country = medal_tally.groupby('country')['total'].sum()
print("Total number of medals won by each country:")
print(total_medals_per_country)
```

```python
# Get the top 10 countries with the highest total number of medals
top_10_countries = total_medals_per_country.sort_values(ascending=False).head(10)

# Visualization
plt.figure(figsize=(12, 8))
top_10_countries.plot(kind='bar', color='gold')
plt.title("Top 10 Countries with the Highest Total Number of Medals")
plt.xlabel("Country")
plt.ylabel("Total Medals")
plt.xticks(rotation=45)
plt.show()
```

### 5. How many different editions of the Olympic Games are included in the dataset?

```python
# Get the top 10 editions with the highest count
top_10_editions = games['edition'].value_counts().head(10)

# Visualization
plt.figure(figsize=(10, 6))
top_10_editions.plot(kind='bar', color='purple')
plt.title("Top 10 Editions of the Olympic Games")
plt.xlabel("Edition")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

### - **Advanced Level** 

### 1. Perform a time series analysis to identify the trend in the number of gold medals won by a specific country over different editions of the Olympics.

```python
# Merging medal_tally with games on 'edition_id'
merged_data = pd.merge(left=medal_tally, right=games, on='edition_id', how='inner')
```

```python
# Display the columns to ensure all necessary columns are present
merged_data.columns
```

```python
# Filter data for a specific country (e.g., USA)
specific_country = 'USA'
country_data = merged_data[merged_data['country_noc_y'] == specific_country] 
```

```python
# Filter data for gold medals only
gold_medals = country_data[['year_x', 'gold']].groupby('year_x').sum().reset_index()
```

```python
# Visualization
plt.figure(figsize=(10, 6))
plt.plot(gold_medals['year_x'], gold_medals['gold'], marker='o', linestyle='-')
plt.title(f"Gold Medals Won by {specific_country} Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Gold Medals")
plt.grid(False)
plt.show()
```

### 2. Correlation between height and weight of athletes in a specific sport (e.g., Swimming)

```python
specific_sport = 'Swimming'
sport_data = athlete_event_result[athlete_event_result['sport'] == specific_sport]
```

```python
# Merge with athlete_bio to get height and weight
sport_athletes = pd.merge(left=sport_data, right=athlete_bio, on='athlete_id', how='inner')
```

```python
# Generate scatter plot for a sample of athletes
sampled_athletes = sport_athletes.sample(n=80, random_state=1)  # Sample 80 athletes for visualization
```

```python
# Scatter plot for height vs weight
plt.figure(figsize=(12, 8))
plt.scatter(sampled_athletes['height'], sampled_athletes['weight'], alpha=0.5)
plt.title(f"Height vs Weight of Athletes in {specific_sport}")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.grid(False)
plt.show()
```

## Geographical Analysis

```python
import geopandas as gpd
```

```python
# Load geographical data
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
# Prepare the data to get the total medals won by each country
medals_by_country = medal_tally.groupby('country_noc')['total'].sum().reset_index()
medals_by_country.columns = ['iso_a3', 'total_medals']
```

```python
# Merge the medals data with the geographical data
world = world.merge(medals_by_country, on='iso_a3', how='left')
```

```python
# Plot the map
plt.figure(figsize=(15, 10))
world.boundary.plot()
world.plot(column='total_medals', cmap='OrRd', legend=True, missing_kwds={'color': 'lightgrey'}, edgecolor='black')
plt.title('Total Medals Won by Each Country')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
